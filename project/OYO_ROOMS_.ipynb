* Rating category distribution

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

### Hyderabad

```python
names=[]
address=[]
city=[]
features=[]
ratings=[]
votes=[]
remarks=[]
prices=[]
distance=[]

for i in range(1,6):
    url="https://www.oyorooms.com/hotels-in-hyderabad/?page="+str(i)
    print(url)

    page=requests.get(url,headers = {'User-Agent':'Mozilla/5.0'})
    soup=BeautifulSoup(page.text)
    
    
    h=soup.find_all("a",class_="c-nn640c u-width100")
    for i in h:
        a=i.text
        names.append(a)
        
    ar=soup.find_all("span",class_="u-line--clamp-2")
    for i in ar:
        a=i.text
        address.append(a.replace(', Hyderabad', ""))
        
    ar=soup.find_all("span",class_="u-line--clamp-2")
    for i in ar:
        a=i.text
        city.append(re.findall(r"Hyderabad",a))
        
    f=soup.find_all("div",class_="amenityWrapper")
    for i in f:
        a=i.text
        features.append(a.split("+")[0])
        
        
    r2=soup.find_all("div",class_="hotelRating")
    for i in r2:
        a=i.text
        ratings.append(a.split()[0])
    
    
    for i in r2:
        a=i.text
        b=re.findall(f"\W(\d+)\sRatings\W",a)
        votes.append(b)
        
    d=soup.find_all("span",class_="listingHotelDescription__distanceText")
    for i in d:
        a=i.text
        distance.append(a)
    

    
    
    p1=soup.find_all("div",class_="listingPrice__numbers")
    for i in p1:
        a=i.text
        prices.append(a.split("â‚¹")[1])
       
   
```

```python
print(len(names))
print(len(address))
print(len(city))
print(len(features))
print(len(ratings))
print(len(votes))
print(len(prices))
print(len(distance))
```

```python
d={"Hotel_Names":names,
    "Distance":distance,
   "Address":address,
   "City":city,
   "Ratings":ratings,
   "Votes":votes,
   "Prices":prices,
   "Features":features
  }
```

```python
df=pd.DataFrame(d)
df
```
